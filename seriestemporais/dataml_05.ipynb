```
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('trends.csv')
data.head()
```

```
data.info()
```

```
data['month'] = pd.to_datetime(data['month'])
data.set_index('month', inplace=True)
```

```
data.info()
```

```
# equivalente ao codigo acima (O efeito dessa transformação é a criação de um novo index do tipo DatetimeIndex.)
data = pd.read_csv('trends.csv', index_col='month', parse_dates=True)
data.info()
```

```
data.plot(figsize=(16,4)) #
```

```
data.diet.rolling(12).mean().plot(figsize=(16,3))
```

```
data.diet.head(15)
```

```
data.diet.groupby(data.index.year).sum()
```

```
data.diet.groupby(data.index.year).sum().plot(figsize=(16,4))
plt.show()
#ou
data.diet.resample('Y').sum().plot(figsize=(16,4))
plt.show()
```

```
data.diet.resample('2Y').sum().plot(figsize=(16,4), linestyle=':', marker='o')
```

```
data.diet.diff().plot(figsize=(16,4))
```

```
data.diet['2008':'2011'].diff().plot(figsize=(16,4))
```

```
data.diet.diff().groupby(data.index.month).sum().plot(figsize=(16,4), kind='bar')
```

```
data.diet.groupby(data.index.month).describe()
```
